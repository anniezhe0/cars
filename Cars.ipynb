# Identify suitable predictors and response variables
A way to identify any suitable predictors and response variables is to use [Bayesian Information Criterion (BIC)](https://www.statology.org/bic-in-python/) and [Akaike Information Criterion (AIC)](https://www.statology.org/aic-in-python/).

```python
import statsmodels.api as sm
#round 1 - BIC
#define response variable
y = cars['price']

#define predictor variables
x = cars[['mileage', 'year']]

#add constant to predictor variables
x = sm.add_constant(x)

#fit regression model
model = sm.OLS(y, x).fit()

#view BIC of model
print(model.bic)
```

```python
#round 2 - BIC
y = cars['price']
x = cars[['mpg', 'engineSize']]
x = sm.add_constant(x)
model = sm.OLS(y, x).fit()
print(model.bic)
```

```python
#round 3 - BIC
y = cars['price']
x = cars[['mpg', 'mileage']]
x = sm.add_constant(x)
model = sm.OLS(y, x).fit()
print(model.bic)
```

```python
#round 4 - BIC
y = cars['price']
x = cars[['engineSize', 'mileage']]
x = sm.add_constant(x)
model = sm.OLS(y, x).fit()
print(model.bic)
```

```python
#round 5 - BIC
y = cars['price']
x = cars[['engineSize', 'mileage', 'mpg']]
x = sm.add_constant(x)
model = sm.OLS(y, x).fit()
print(model.bic)
```

```python
#round 6 - BIC
y = cars['tax']
x = cars[['engineSize', 'mileage', 'mpg']]
x = sm.add_constant(x)
model = sm.OLS(y, x).fit()
print(model.bic)
```

```python
#round 7 - BIC
y = cars['mpg']
x = cars[['engineSize', 'mileage']]
x = sm.add_constant(x)
model = sm.OLS(y, x).fit()
print(model.bic)
```

Out of all rounds, it appears the 7th round is smaller one, so I'll select the response variable and predictors variables from that round.

```python
#round 1 - AIC
y = cars['mpg']
x = cars[['mileage', 'engineSize', 'year']]
x = sm.add_constant(x)
model = sm.OLS(y, x).fit()
print(model.aic)
```

```python
#round 2 - AIC
y = cars['mpg']

#define predictor variables
x = cars[['engineSize', 'mileage']]

#add constant to predictor variables
x = sm.add_constant(x)

#fit regression model
model = sm.OLS(y, x).fit()

#view BIC of model
print(model.aic)
```

```python
#round 3 - AIC
y = cars['mileage']

#define predictor variables
x = cars[['engineSize', 'mpg']]

#add constant to predictor variables
x = sm.add_constant(x)

#fit regression model
model = sm.OLS(y, x).fit()

#view AIC of model
print(model.aic)
```

My final selection: first round from AIC because the number for AIC is smaller. 

# Polynomial Regression

```python
from sklearn.preprocessing import PolynomialFeatures

#setting polynomial degree = 2
polynomial_converter = PolynomialFeatures(degree=2,include_bias=False)
y = cars['mpg']
X = cars[['mileage', 'engineSize', 'year']]

poly_features = polynomial_converter.fit_transform(X)
poly_features.shape
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(poly_features, y, test_size=0.3, random_state=101)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression(fit_intercept=True).fit(X_train,y_train)
```

```python
test_predictions = model.predict(X_test)
from sklearn.metrics import mean_absolute_error,mean_squared_error
MAE = mean_absolute_error(y_test,test_predictions)
MSE = mean_squared_error(y_test,test_predictions)
RMSE = np.sqrt(MSE)

print("MAE: ", MAE)
print("MSE: ", MSE)
print("RMSE", RMSE)
```

My interpretation of the result:
- Mean absolute error (MAE) is 8.41, which is our calculated eror between actual values and predicted values. 
- Mean squared error (MSE) is 218.92, which means there's large difference between actual values and predicted values
- Root mean squared error (RMSE) is 14.79, which means the average error rate is about 14%

Outcome: I need to revise my approach to handling the data in order to obtain a lower MSE and RMSE.
